# FLUX-PIHM 
- Coupled LSM and WS 
- Flux: based on NOAH MP LSM (hey Liang!!)
    - 1D vertical 
- PIHM: Penn State Integrated Hydrologic Model
    - 2.5D (horizontal, shallow subsurface) watershed
- Model domain: 2.5 D mostly horizontal watershed domain with only 2 vertical layers 
    - Unstructured triangular cells 
## Initial Conditions
- soil moisture, head

## Model engine
- 2D/1D St. Venant eqn for overland / channel flow
- GW flow eqn
- Richards eqn for vertical UZ flow
- Energy balance eqn
- Solution: finite volume

## Parameters
- Topography
    - Land cover classes
        - Preset or customize with parameters
- Soil parameters
- Plant 

## Boundary conditions
- Weather forcing (precip, temp, radiation, air P)
- No flow at WS boundary

## Watershed Discretization
- AKA catchment
- Entire area where water will drain to a common discharge / outlet 
- Does not align well with GW frameworks 
    - Aquifer systems are independent from surface systems 
- Unlike GW, we can't choose an arbitrary area, we must define the watershed
    - Use PIHM GIS 
- Triangulated Irregular Network (TIN)
    - Good for efficient flow calculations, esp over steep terrain 

## Other watershed models
- Parflow: Full 3D process-based model
- MIKE-SHE: 3D integrated (proprietary)
- SWAT-MODFLOW
- GS-FLOW 
    - GSFLOW-GRASS pre and post processing


## Generating Model Domain
- Need Digital Elevation Map (DEM)
1. Use DEM in QGIS 3.xx to convert DEM to .asc 
2. Use PIHMGIS to create TIN mesh
